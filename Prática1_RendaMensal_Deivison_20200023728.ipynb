**Atividade 2:**

Implemente o algoritmo de Regressão Linear que retorne o vetor de pesos $w=(X^TX)^{-1}X^Ty$

```python
class Linear_Regression:
    def fit(self, _X, _y):
        xinv = np.linalg.inv(_X.T.dot(_X))
        self.w = xinv.dot(_X.T).dot(_y)      

    def predict(self, _x):
        return [self.w.T.dot(i) for i in _x]
     
    def getW(self):
        return self.w
```
