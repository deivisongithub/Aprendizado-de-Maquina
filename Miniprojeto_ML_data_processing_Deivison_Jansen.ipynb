```python
#Lendo arquivos csv do treino e do test
train = pd.read_csv("data/train.csv",sep=",")
test = pd.read_csv("data/test.csv",sep=",")

#Removendo coluna sem utilidade
train = train.drop(columns="Unnamed: 0")
test = test.drop(columns="Unnamed: 0")
```

```python
#Retirando os labels para processamento dos dados
train = np.array(train.drop(columns="label"))
test = np.array(test.drop(columns="label"))
```

###pré-processamento e dataset definitivo

Para reduzir a dimensão das amostras vamos utilizar os conceitos de **intensidade** e **Simetria**

```python
#Cálculo da intensidade de cada número
def get_intensity(data):
  intensity = []
  for i in tqdm(range(data.shape[0])):
    intensity.append(np.sum(data[i])/255)
  return intensity
```

```python
#Cálculo da simetria vertical de cada número
def get_vertical_symmetry(data):
  symmetry = []
  for line in tqdm(range(data.shape[0])):
    matrix = data[line].reshape(28,28)
    line_image = 0
    for i in range(28):
      for j in range(14):
        line_image += np.sum(abs(matrix[i][j] - matrix[i][27 - j]))
    line_dataset = line_image / 255
    symmetry.append(line_dataset)
  return symmetry
```

```python
#Cálculo da simetria horizontal de cada número
def get_horizontal_symmetry(data):
  symmetry = []
  for line in tqdm(range(data.shape[0])):
    matrix = data[line].reshape(28,28)
    line_image = 0
    for i in range(14):
      for j in range(28):
        line_image += np.sum(abs(matrix[i][j] - matrix[j][27 - i]))
    line_dataset = line_image / 255
    symmetry.append(line_dataset)
  return symmetry
```

```python
#Reduzindo a dimensão dados de treino e teste

#Treino
intensity_train = get_intensity(train)
symmetry_v_train = get_vertical_symmetry(train)
#symmetry_h_train = get_horizontal_symmetry(train)


#Teste
intensity_test = get_intensity(test)
symmetry_v_test = get_vertical_symmetry(test)
#symmetry_h_test = get_horizontal_symmetry(test)

#Criando os novos Dataset com dimensões reduzidas

#Treino
new_train = pd.DataFrame(label_train)
new_train['intensity'] = intensity_train
new_train['symmetry'] = symmetry_v_train

#Teste
new_test = pd.DataFrame(label_test)
new_test['intensity'] = intensity_test
new_test['symmetry'] = symmetry_v_test
```
